# Project Airline Assistant

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
load_dotenv(override=True)
openai_api_key=os.getenv('OPENAI_API_KEY')
if openai_api_key:
    print(f"The API KEY starts with {openai_api_key[:8]}")
else:
    print(f"TheAPI Key doesnot exists")
```

```python
openai=OpenAI()
MODEL="gpt-4.1-mini"
```

```python
system_message="""you are helpful assistant for flight booking called FlightAI.
                Give a short,courteous answers not more than 1 sentence.
                Always be accurate,if you dont know the answer then say no."""
```

```python
def chat(message,history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    messages = [{"role":"system", "content": system_message}] + history + [{"role":"user", "content": message}]
    response = openai.chat.completions.create(model=MODEL, messages=messages)
    return response.choices[0].message.content

gr.ChatInterface(fn=chat,type="messages").launch()
```
