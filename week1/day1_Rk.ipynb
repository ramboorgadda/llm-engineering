```python
system_prompt = """
You are a good assistant that analyzes the contents of a website,
and provides good summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

```python
user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes academic or announcements, then summarize these too.

"""
```

```python
messages=[{"role":"system","content":"you are a good assistant"},
{"role":"user","content":"add 2 numbers 3 + 4"}]
response=openai.chat.completions.create(model="gpt-5-nano",messages=messages)
response.choices[0].message.content
```

```python
# See how this function creates exactly the format above

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

```python
messages_for("https://ngit.ac.in/management-2/")
```

```python
def summarize(url):
    website=fetch_website_contents(url)
    messages=messages_for(website)
    response=openai.chat.completions.create(model="gpt-5-nano",messages=messages)
    return response.choices[0].message.content
```

```python
summarize("https://ngit.ac.in/management-2/")
```

```python
def display_markdown(url):
    summary=summarize(url)
    display(Markdown(summary))
```

```python
display_markdown("https://ngit.ac.in/management-2/")
```
